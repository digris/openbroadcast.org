```python
from elasticsearch import Elasticsearch

#from django_elasticsearch_dsl import Index

#index = Index('releases')



client = Elasticsearch('10.35.30.231:9200')

response = client.search(
    index="releases",
    body={
        "aggs": {
            "duplicates": {
                "terms": {
                    "script": "doc['name.raw'].value + ' ' + doc['artist_display.raw'].value",
                    "size": 10000,
                    "min_doc_count": 2
                },
                "aggs": {
                    "duplicates": {
                        "top_hits": {
                            "_source": {
                                "includes": [
                                    "id",
                                    "name",
                                    "artist_display"
                                ]
                            },
                            "size": 20
                        }
                    }
                }
            }
        }
    }
)


print(response['aggregations']['duplicates']['buckets'])

```
